a) Visualização de intervalo de dados em modo texto: a partir de entradas do usuário, sua implementação deve permitir a visualização dos dados que foram carregados do arquivo. O usuário deve informar o período que quer ver, ou seja, deve indicar o mês e ano iniciais, bem como o mês e ano finais que deseja visualizar os dados. Permita também que o usuário informe se quer ver 1) todos os dados, 2) apenas os de precipitação, 3) apenas os de temperatura, ou 4) apenas os de umidade e vento para o período informado. 

```python
#Implementar com funções.

def opcao_invalida():
    print('Opção inválida! Aperte qualquer tecla para voltar ao início.')
    main()


def escolha(): #provavelmente ce vai ter que fazer de uma forma diferente, vê de noite.
    while True:
        try:
            mesinicial = int(input('Informe o valor do mês inicial para visualizar (1 a 12):'))
            if 1 <= mesinicial <= 12:
                return mesinicial
            else:
                opcao_invalida()
        except ValueError:
            opcao_invalida()
        break

    while True:
        try:
            mesfinal = int(input('informe o valor do mês final para visualizar (1 a 12):'))
            if 1 <= mesfinal <= 12:
                return mesfinal
            else:
                opcao_invalida()
        except ValueError:
            opcao_invalida()
        break

    while True:
        try:
            anoinicial = int(input('Informe o valor do ano inicial para visualizar (1961 a 2016):'))
            if 1961 <= anoinicial <= 2016:
                return anoinicial
            else:
                opcao_invalida()
        except ValueError:
            opcao_invalida()
        break

    while True:
        try:
            anofinal = int(input('informe o valor do ano final para visualizar (1961 a 2016):'))
            if 1961 <= anofinal <= 2016:
                return anofinal
            else:
                opcao_invalida()
        except ValueError:
            opcao_invalida()
        break

mesinicial = escolha()
mesfinal = escolha()
anoinicial = escolha()
anofinal  = escolha()

print(mesinicial)
print(mesfinal)
print(anoinicial)
print(anofinal)
def main():
    os.system('cls')
    escolha()

if __name__ =='__main__ ':
    main()
```
